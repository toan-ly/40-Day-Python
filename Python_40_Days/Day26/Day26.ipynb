```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Question 1
im = cv2.imread('dataset/1.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 9))

# Question 1 
plt.subplot(221)
plt.imshow(im)
plt.title('Original Image')
plt.axis('off')

# Question 2
plt.subplot(222)
im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
plt.imshow(im_gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

# Question 3
plt.subplot(223)
im_increase = im.astype(np.float32) + 50
im_increase = np.clip(im_increase, 0, 255)
im_increase = im_increase.astype(np.uint8)
plt.imshow(im_increase)
plt.title('Increased by 50')
plt.axis('off')

plt.subplot(224)
im_decrease = im.astype(np.float32) - 80
im_decrease = np.clip(im_decrease, 0, 255)
im_decrease = im_decrease.astype('uint8')
plt.imshow(im_decrease)
plt.title('Decreased by 80')
plt.axis('off')

plt.tight_layout()
plt.show()
```
