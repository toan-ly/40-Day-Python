```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

filename = librosa.example('nutcracker')
audio, sr = librosa.load(filename)

time = np.linspace(0, audio.shape[0] / sr, num=audio.shape[0])
plt.plot(time, audio, color='blue')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (quantized)')
plt.title('Wav file visualization')
plt.axhline(y=0, color='r', linestyle='-')
plt.show()
```

```python
import IPython.display as ipd
from gtts import gTTS
import librosa
import numpy as np
import IPython

lang = 'vi'
output_filename = 'record.mp3'
content = 'Xin chào mọi người'
output = gTTS(content, lang=lang, slow=False)

output.save(output_filename)
data, sr = librosa.load(output_filename)

ipd.display(ipd.Audio(np.transpose(data), rate=sr))
```

```python
from scipy.fft import fft

test_case_1 = 'test.mp3'
audio, sr = librosa.load(test_case_1)

plt.figure(figsize=(12, 6))
plt.subplot(211)
plt.plot(audio)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Original Audio Signal')

plt.subplot(212)
fft_output = fft(audio)
plt.plot(fft_output)
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.title('Audio Signal after Fast Fourier Transform')

plt.tight_layout()
plt.show()
```
